```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.inspection import permutation_importance
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# Load your dataset
heart_data = pd.read_csv('heart.csv')

# Check for null values
null_values = heart_data.isnull().sum()
print("Null Values in Each Column:")
print(null_values)


duplicate_val=heart_data[heart_data.duplicated()]#checking for duplicate values

if duplicate_val.empty:
    print("There are no duplicate values.")
else:
    print("There are duplicate values present.")
duplicate_val
```

```python

# Data Preprocessing
label_encoders = {}
for col in heart_data.select_dtypes(include=['object']).columns:
    encoder = LabelEncoder()
    heart_data[col] = encoder.fit_transform(heart_data[col])
    label_encoders[col] = encoder

scaler = StandardScaler()
heart_data[['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']] = scaler.fit_transform(heart_data[['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']])

```

```python
# Split Data
features = heart_data.drop('HeartDisease', axis=1)
target = heart_data['HeartDisease']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Logistic Regression Model
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
predictions_logistic = logistic_model.predict(X_test)

# Evaluation
print("Logistic Regression Model")
print(classification_report(y_test, predictions_logistic))

# Confusion Matrix
conf_matrix_logistic = confusion_matrix(y_test, predictions_logistic)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_logistic, annot=True, fmt='d', cmap='Blues', xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Feature Importance Visualization
importance_df = pd.DataFrame({
    'Feature': features.columns,
    'Coefficient': logistic_model.coef_[0]  # Get the coefficients for the features
})

# Sort the DataFrame by absolute value of coefficients
importance_df['Absolute Coefficient'] = importance_df['Coefficient'].abs()
importance_df = importance_df.sort_values(by='Absolute Coefficient', ascending=False)

# Visualization of Feature Importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Absolute Coefficient', y='Feature', data=importance_df, palette='viridis')
plt.title('Feature Importance using Logistic Regression Coefficients')
plt.xlabel('Absolute Coefficient')
plt.ylabel('Features')
plt.show()
```

```python
# Random Forest Model
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train, y_train)

# Evaluation
predictions = random_forest_model.predict(X_test)
print("Random Forest Model")
print(classification_report(y_test, predictions))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, predictions)
print("Confusion Matrix:\n", conf_matrix)
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.2f}")

# Visualization of Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.title('Confusion Matrix for Random Forest Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Feature Importance Visualization
importance_df = pd.DataFrame({
    'Feature': features.columns,
    'Importance': random_forest_model.feature_importances_  # Get feature importances
})

# Sort the DataFrame by importance
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Visualization of Feature Importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
plt.title('Feature Importance using Random Forest')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()
```

```python
# SVM Model
svm_model = SVC(probability=True)  # Enable probability estimates
svm_model.fit(X_train, y_train)

# Evaluation
predictions = svm_model.predict(X_test)
print("SVM Model")
print(classification_report(y_test, predictions))

# Confusion Matrix
matrix = confusion_matrix(y_test, predictions)
print("Confusion Matrix:\n", matrix)
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.2f}")

# Visualization of Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.title('Confusion Matrix for SVM Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Compute permutation importance
results = permutation_importance(svm_model, X_test, y_test, n_repeats=30, random_state=42)

# Create a DataFrame for feature importances
importance_df = pd.DataFrame({
    'Feature': features.columns,
    'Importance': results.importances_mean
})

# Sort the DataFrame by importance
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Visualization of Feature Importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
plt.title('Feature Importance using Permutation Importance for SVM Model')
plt.xlabel('Mean Importance')
plt.ylabel('Features')
plt.show()

```
